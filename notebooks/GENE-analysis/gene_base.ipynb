```python
file=open("C:\\Users\\giann\\OneDrive\\Desktop\\base_gene.txt")
next(file)
ky = []
gamma = []
omega = []
for columns in (raw.strip().split() for raw in file):  
        ky.append(float(columns[0]))
        gamma.append(float(columns[1]))
        omega.append(float(columns[2]))
```

```python
omega_plfr = -2496238.5965949353
krho_sim = 0.246
```

```python
fig,ax = plt.subplots()
ax.set_title('Base case, $\psi=0.97$')
ax.plot(ky,omega,"^-", linewidth=1.0)
ax.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.1e'))
ax.set_xlabel(r'$k_y \rho_i$')
ax.set_ylabel(r'$\omega \left(\frac{rad}{sec}\right)$')
ax.plot(krho_sim, omega_plfr,'bo',markersize=12)
ax.axvline(x=krho_sim,color='b',linewidth=0.7)
plt.show()
```
